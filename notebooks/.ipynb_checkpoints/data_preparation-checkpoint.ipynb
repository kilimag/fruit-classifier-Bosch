```python
import pandas as pd

# Datei laden
file_path = "../data/fruit_data.xlsx"  # Ersetze mit dem tatsächlichen Pfad zur Datei
data = pd.read_excel(file_path)

# Die ersten Zeilen des Datensatzes anzeigen
print(data.head())

# Informationen über die Daten anzeigen
print(data.info())
```

```python
# Gesamtanzahl der fehlenden Werte
total_missing = data.isnull().sum().sum()
print(f"Gesamtanzahl der fehlenden Werte im Datensatz: {total_missing}")
```

```python
import matplotlib.pyplot as plt

# Häufigkeiten fruit_type berechnen
fruit_type_counts = data["fruit_type"].value_counts()
print(fruit_type_counts)



```

```python
#Fruit Balkendiagramm

# Balkendiagramm
fruit_type_counts.plot(kind="bar", color="skyblue", figsize=(8, 6))

# Titel und Achsenbeschriftungen
plt.title("Häufigkeiten der Kategorien in der Variable 'fruit_type'", fontsize=16)
plt.xlabel("Kategorie", fontsize=14)
plt.ylabel("Häufigkeit", fontsize=14)

# Diagramm anzeigen
plt.xticks(rotation=0)  # Kategorien horizontal anzeigen
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: Gitterlinien
plt.show()
```

```python
# Häufigkeiten für color berechnen
color_counts = data["color"].value_counts()
print(color_counts)
```

```python

#Color Balkendiagramm

# Balkendiagramm
color_counts.plot(kind="bar", color="skyblue", figsize=(8, 6))

# Titel und Achsenbeschriftungen
plt.title("Häufigkeiten der Kategorien in der Variable 'fruit_type'", fontsize=16)
plt.xlabel("Kategorie", fontsize=14)
plt.ylabel("Häufigkeit", fontsize=14)

# Diagramm anzeigen
plt.xticks(rotation=0)  # Kategorien horizontal anzeigen
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: Gitterlinien
plt.show()
```

```python
#Yellow1 zu Yellow machen. Offensichtlicher Schreibfehler.

# Mapping für die Zusammenführung
color_mapping = {
    "Yellow1": "Yellow",
}

# Mapping anwenden
data["color"] = data["color"].replace(color_mapping)

# Ergebnis anzeigen
print(data["color"].value_counts())
```

```python
#Pale Yellow genauer betrachten:
import matplotlib.pyplot as plt
import seaborn as sns

# Zählen der Häufigkeit von Kombinationen aus 'Color' und 'Fruit'
color_fruit_counts = data.groupby(['color', 'fruit_type']).size().reset_index(name='Count')

# Gestapeltes Balkendiagramm erstellen
plt.figure(figsize=(10, 6))

# Mit seaborn barplot (gestapelt durch hue)
sns.barplot(x="color", y="Count", hue="fruit_type", data=color_fruit_counts, palette="Set2")

# Titel und Achsenbeschriftungen
plt.title("Verteilung von Früchten pro Farbe", fontsize=14)
plt.xlabel("Farbe", fontsize=12)
plt.ylabel("Anzahl der Beobachtungen", fontsize=12)

# Zeigen des Plots
plt.show()
```

```python
# Pale Yellow zu Yellow?
# Nach Abwägung der Vor- und Nachteile habe ich mich entschieden, die Kategorie "Pale Yellow" beizubehalten. 
# Die Gründe hierfür sind:

# 1. Möglicher informativer Unterschied:
# "Pale Yellow" könnte eine spezifische Bedeutung haben, beispielsweise ein bestimmtes Reifestadium der Bananen. 
# Dies könnte ein relevanter Faktor für die Klassifikation sein, der durch die Zusammenfassung verloren gehen würde.

# 2. Relevanz für die Interpretation:
# Die Unterscheidung zwischen "Pale Yellow" und "Yellow" könnte für spätere Analysen oder Interpretationen von Bedeutung sein, insbesondere wenn die Farbe als Indikator für Qualität oder Reife genutzt werden soll.

# 3. Ausreichend Daten für Robustheit:
# Trotz der geringen Anzahl von 9 Beobachtungen in der Kategorie "Pale Yellow" sind die anderen Farben und Klassen gut vertreten, sodass die Modelle robust genug sein sollten, mit dieser kleinen Kategorie umzugehen.

# Durch die Beibehaltung dieser Kategorie gehen keine potenziell relevanten Informationen verloren.
```

```python
# Häufigkeiten size berechnen
size_counts = data["size"].value_counts()
print(size_counts)
```

```python
#Largee zu Large hinzufügen, da Schreibfehler.
# Mapping für die Korrektur
size_mapping = {
    "Largee": "Large",  # Tippfehler korrigieren
}

# Mapping anwenden
data["size"] = data["size"].replace(size_mapping)

# Ergebnis anzeigen
print(data["size"].value_counts())
```

```python
#Size Rangordnung
# Kategorien definieren
size_categories = ["Tiny","Small", "Medium", "Large"]

# Größe als kategorische Variable mit Reihenfolge festlegen
data["size"] = pd.Categorical(data["size"], categories=size_categories, ordered=True)

# Ergebnis anzeigen
print(data["size"].value_counts())
```

```python
#weight 

# Histogramm
plt.figure(figsize=(8, 6))
plt.hist(data["weight"].dropna(), bins=10, color="skyblue", edgecolor="black")
plt.title("Verteilung von 'weight'", fontsize=16)
plt.xlabel("Weight", fontsize=14)
plt.ylabel("Häufigkeit", fontsize=14)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Anzahl der Beobachtungen mit 'weight' > 200
count_above_200 = data[data["weight"] > 200].shape[0]
print(f"Anzahl der Beobachtungen mit 'weight' > 200: {count_above_200}")
```

```python
#Ausreiser von weight aus dem Datensatz nehmen
#Da wir nur 200 Beobachtungen haben, könnten diese 2 die Gesamtanaylse verzerren.
#Modelle(v.a Klassifikatoren) würden darunter leiden.

# Entfernen der Beobachtungen mit 'weight' > 200
data_cleaned = data[data["weight"] <= 200]
```

```python
# Histogramm nach Entfernen der Ausreißer
plt.figure(figsize=(8, 6))
plt.hist(data_cleaned["weight"], bins=10, color="skyblue", edgecolor="black")
plt.title("Verteilung von 'weight' nach Entfernen der Ausreißer")
plt.xlabel("Weight")
plt.ylabel("Häufigkeit")
plt.show()
```
