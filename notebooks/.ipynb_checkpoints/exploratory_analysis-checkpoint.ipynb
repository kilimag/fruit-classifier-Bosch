```python
import pandas as pd

# Datei laden
file_path = "../data/fruit_data_cleaned.xlsx"  # Ersetze mit dem tatsächlichen Pfad zur Datei
data = pd.read_excel(file_path)

# Die ersten Zeilen des Datensatzes anzeigen
print(data.head())

# Informationen über die Daten anzeigen
print(data.info())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(x='fruit_type', y='weight', data=data)
```

```python
#Gewichtsunterschiede zwischen den Früchten:

#Trauben haben deutlich geringere Gewichtswerte als Äpfel und Bananen.
#Äpfel und Bananen haben vergleichbare Gewichtsspektren, wobei Äpfel tendenziell etwas schwerer sind.

#Ausreißer:

#Es gibt Ausreißer. Zum Beispiel:
#Ein niedriger Wert bei Äpfeln.
#Einige Werte bei Trauben außerhalb der regulären Verteilung.


# Filter nur für Äpfel
apple_data = data[data['fruit_type'] == 'apple']

# Histogramm
plt.figure(figsize=(8, 6))
sns.histplot(apple_data['weight'], bins=10, kde=True, color='red')
plt.title("Gewichtsverteilung der Äpfel")
plt.xlabel("Gewicht")
plt.ylabel("Anzahl")
plt.show()

```

```python

banana_data = data[data['fruit_type'] == 'banana']

# Histogramm
plt.figure(figsize=(8, 6))
sns.histplot(banana_data['weight'], bins=10, kde=True, color='yellow')
plt.title("Gewichtsverteilung der Bananen")
plt.xlabel("Gewicht")
plt.ylabel("Anzahl")
plt.show()
```

```python
#Boxplot size und weight
sns.boxplot(x='size', y='weight', data=data)
```

```python
#Trennbarkeit der Größenklassen:

#Die Tiny-Kategorie unterscheidet sich klar von den anderen durch ihr deutlich geringeres Gewicht.
#Small-, Medium- und Large-Kategorien überschneiden sich teilweise, was darauf hinweist, dass die Größe allein möglicherweise kein perfektes Kriterium zur Unterscheidung ist.
```

```python
# Kreuztabelle erstellen
heatmap_data = pd.crosstab(data['size'], data['fruit_type'])

# Heatmap plotten
sns.heatmap(heatmap_data, annot=True, cmap='Blues')
plt.title('Häufigkeit von size pro fruit_type')
plt.xlabel('Fruchttyp')
plt.ylabel('Größe')
plt.show()
```

```python
# Kreuztabelle erstellen
heatmap_data = pd.crosstab(data['color'], data['fruit_type'])

# Heatmap plotten
sns.heatmap(heatmap_data, annot=True, cmap='Blues')
plt.title('Häufigkeit von size pro fruit_type')
plt.xlabel('Fruchttyp')
plt.ylabel('Farbe')
plt.show()
```
